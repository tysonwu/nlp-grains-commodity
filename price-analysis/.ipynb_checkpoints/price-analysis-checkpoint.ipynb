# Price-sentiment Analysis

- Price data obtained from Quandl https://blog.quandl.com/api-for-commodity-data

## Obtaining price data

```python
"""
http://www.quandl.com/api/v3/datasets/WORLDBANK/WLD_TEA_MOMBASA
Barley, Western Canada (ODA) ODA ODA/PBARL_USD
Barley, Western Canada (WB) WORLDBANK WORLDBANK/WLD_BARLEY
Corn, Top Flight Grain Co-op TFGRAIN TFGRAIN/CORN
Corn, CBOT Corn Futures CME CHRIS/CME_C1
Corn, U.S. Gulf Imports WORLDBANK WORLDBANK/WLD_MAIZE
Oats, CBOT Oats Futures CME CHRIS/CME_O1
Rice, Thailand (ODA) ODA ODA/PRICENPQ_USD
Rice, CBOT Rice Futures CME CHRIS/CME_RR1
Rice, Top Flight Grain Co-op TFGRAIN TFGRAIN/SOYBEANS
Soybean, CBOT Soybean Futures CME CHRIS/CME_S1
Soybean, CBOT Soybean Meal Futures CME CHRIS/CME_SM1
Wheat, Soybean Oil WORLDBANK WORLDBANK/WLD_SOYBEAN_OIL
Wheat, CBOT Soybean Oil Futures CME CHRIS/CME_BO1
Wheat, Gulf of Mexico ODA ODA/PWHEAMT_USD
Wheat, CBOT Wheat Futures CME CHRIS/CME_W1
"""

import requests
import time
import pandas as pd
import numpy as np
from datetime import datetime

# using CME price
api_params = {
    'corn': 'CHRIS/CME_C1',
    'oats': 'CHRIS/CME_O1',
    'rice': 'CHRIS/CME_RR1',
    'soybean': 'CHRIS/CME_S1',
    'wheat': 'CHRIS/CME_W1'
}


def get_price(param):
    
    url = f'http://www.quandl.com/api/v3/datasets/{param}'
    r = requests.get(url).json()
    time.sleep(0.5)
    colnames = r['dataset']['column_names']
    data = r['dataset']['data']
    df = pd.DataFrame(data, columns=colnames)
    df['Date'] = df['Date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
    df = df.sort_values('Date')
    df['logreturn'] = np.log(df['Last']/df['Last'].shift(1))
    
    df = df.reset_index()
    df = df.set_index('Date')
    
    return df


prices = {}

for k, v in api_params.items():
    prices[k] = get_price(v)
```

## Plot historical price data

```python
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [30,8]

for k in api_params.keys():
    plt.plot(prices[k].index, prices[k]['Last'], label=k)
plt.legend()
plt.xlim([datetime(2015,1,1),datetime(2020,11,30)])
plt.yscale('log')
plt.show()
```

## Load processed news and tweets data

```python
import pandas as pd
from datetime import datetime

KEYWORDS = ['corn','oats','rice','soybean','wheat','grains']

def load_data():
    df = pd.read_csv('./dataframe.csv',lineterminator='\n')
    return df


def preprocessing(df):
    df = df[['date','query','lexicon_score']]
    df['date'] = df['date'].apply(lambda x: x[:11].strip())
    df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
    
    dfs = {}
    for key in KEYWORDS:
        # aggregate data by time
        temp = df[df['query']==key]
        dfs[key] = temp[['date','lexicon_score']].groupby('date').mean()
        
    return dfs


df = load_data()
print(df.head())
dfs = preprocessing(df)
```

## Daily sentiment graph

- This shows a noisy data in daily aggregation

```python
# daily sentiment graph

import matplotlib.pyplot as plt
import ta
import numpy as np

query = 'corn'
plt.rcParams["figure.figsize"] = [12,8]

df = dfs[query]
print(df)
    
pricedf = prices[query][['logreturn']]

plt.plot(df.index, df['lexicon_score'])
plt.xlim(datetime(2015,1,1), datetime(2020,10,31))
plt.show()
```

## 3-month aggregated sentiment graph

```python
import matplotlib.pyplot as plt
import ta
import numpy as np

plt.rcParams["figure.figsize"] = [12,8]

df = dfs[query]
df = df.resample('3M').mean()
print(df)

lexicon_score_sma = ta.trend.SMAIndicator(close=df['lexicon_score'], n=8)
df['lexicon_score_sma'] = lexicon_score_sma.sma_indicator().ewm(alpha=0.6).mean()
    
plt.plot(df.index, df['lexicon_score'])

pricedf = prices[query][['logreturn']]
pricedf = pricedf.resample('3M').sum()

plt.plot(pricedf.index, pricedf['logreturn'])
plt.xlim(datetime(2015,1,1), datetime(2020,10,31))

plt.show()
```

# Linear regression to lagged sentiment scores

- With various parameter settings

```python
from sklearn import linear_model
import statsmodels.api as sm


# KEYWORDS = ['corn','oats','rice','soybean','wheat']
KEYWORDS = ['corn','rice','soybean','wheat','rice']

for query in KEYWORDS:
    print(query)
#     dfm = dfs[query].join(prices[query]['logreturn'], how='left')
    dfm = dfs[query].join(prices[query]['logreturn'], how='left')
    dfm = dfm.resample('2M').agg({'lexicon_score':'mean','logreturn':'sum'})
    dfm = dfm[dfm.index >= datetime(2017,1,1)]

    for n in range(1, 7):
        dfm[f'lexicon_score_lag_{n}'] = dfm['lexicon_score'].shift(n)

    dfm = dfm.dropna()

    X = dfm.drop(['logreturn','lexicon_score'], axis=1)
    Y = dfm[['logreturn']]

    # with sklearn
    regr = linear_model.LinearRegression()
    regr.fit(X, Y)

    print('Intercept: \n', regr.intercept_)
    print('Coefficients: \n', regr.coef_)


    # with statsmodels
    X = sm.add_constant(X) # adding a constant

    model = sm.OLS(Y, X).fit()
    print_model = model.summary()
    print(print_model)
    print('###################################################')
```
